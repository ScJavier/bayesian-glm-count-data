```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

np.random.seed(1010)
```

# Simula datos

```python
lambdas = [0.1, 0.5, 1.0, 5.0, 10.0, 50.0]
n_sample = 1000
simulated_values = {}

for lam in lambdas:
    simulated_values[lam] = np.random.poisson(lam, size=n_sample)
```

# Visualiza datos

```python
fig, ax = plt.subplots(2, 3, figsize=(12, 8))
for index, lam in enumerate(lambdas):
    i = index // 3
    j = index % 3
    sns.histplot(x=simulated_values[lam], color='darkcyan', ax=ax[i, j])
    ax[i,j].set_title(f'Possion sample - $\lambda$ = {lam}')
plt.tight_layout()
plt.show()
```
