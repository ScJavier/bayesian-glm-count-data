# Modelos para Datos de Conteo: Un Recorrido Bayesiano

*Horseshoe Crab Satellites — Agresti (2002)*

## Origen del problema

En un problema aplicado de predicción de demanda, la regresión Poisson fue el primer
modelo candidato. Su supuesto central — que la varianza de la respuesta es igual a su
media (Var[Y] = E[Y]) — rara vez se cumple en datos reales. La sobredispersión llevó a
explorar Binomial Negativa, y la abundancia de ceros a modelos Zero-Inflated. Este
minicurso documenta ese recorrido de forma sistemática.

El vehículo de estudio es el dataset **Horseshoe Crab Satellites** (Agresti 2002): un
ejemplo clásico donde los problemas son concretos y verificables, y el comportamiento
de cada modelo es interpretable.

## El dataset

**Horseshoe Crab Satellites** (n = 173 hembras, Agresti 2002):

| Variable | Tipo | Descripción |
|---|---|---|
| `Sa` | Conteo | Número de satélites (variable respuesta) |
| `W` | Continua | Ancho de caparazón (cm); predictor principal |
| `Wt` | Continua | Peso (kg); alta colinealidad con W |
| `C` | Ordinal 1–4 | Color del caparazón (oscuro → claro) |
| `S` | Ordinal 1–3 | Condición de la espina (buena → desgastada) |

Características que motivan el minicurso:

| Estadístico | Valor |
|---|---|
| Media de Sa | 2.92 |
| Varianza de Sa | 9.91 |
| Índice de dispersión D = Var/Media | **3.38** |
| Proporción de ceros | **≈33%** |

La sobredispersión (D >> 1) invalida el modelo Poisson. El exceso de ceros motiva los
modelos Zero-Inflated.

## Pregunta de investigación

> ¿Cómo modelar datos de conteo con sobredispersión y exceso de ceros?  
> ¿Qué gana el paradigma bayesiano sobre el frecuentista en este contexto?

La respuesta no es "usa Stan en lugar de statsmodels" — es entender qué dimensiones
del ajuste captura cada modelo y cómo verificarlo mediante el workflow bayesiano
(Gelman et al. 2020).
